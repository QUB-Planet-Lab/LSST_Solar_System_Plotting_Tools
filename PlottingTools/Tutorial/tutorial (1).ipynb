Plotting functions with their full parameters

This gives you access to the queries, so if you only want a dataframe to manipulate you can just query using this function.
Queries(startdate=None, enddate=None, mindistance=None, maxdistance=None, Query=None, a_min=None, a_max=None,
            q_min=None, q_max=None, i_min=None,
            i_max=None, e_min=None, e_max=None)

These are the columns available with each Query type.

```python
for QueryNum in range(1,9):
    print([ item for item in Queries(startdate=0, enddate=0, mindistance=None, maxdistance=None, Query=QueryNum, a_min=None, a_max=None,
            q_min=None, q_max=None, i_min=None,
            i_max=None, e_min=0, e_max=0).keys()  ])
```
