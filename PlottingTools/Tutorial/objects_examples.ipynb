# Objects examples

The Objects class aims to provide access to unique observations made by LSST within a certain time range and specified orbital parameters.

```python
import sys

sys.path.append("../Source")
```

### Object class

```python
from collection import Objects
```

Objects(

    start_time,
    end_time, 
    **orbital_elements
):

```python
%%time

#instatiate Objects
example_objs = Objects(
    start_time = "2023-04-09",
    end_time = "2023-04-12",
    min_a = 0,
    max_a = 2,
    max_e = 0.8
)
```

### Plot objects

plot_objects(

        filters: Optional[list] = None,
        title : Optional[str] = None,
        time_format: Optional[Literal['ISO', 'MJD']] = 'ISO',
        projection: Optional[Literal['2d', '3d']] = '2d',
        library: Optional[str] =  "seaborn",
        cache_data: Optional[bool] = False
): 

```python
%%time
obj_plot = example_objs.plot_objects(
    title = "Objects detected between 09-12 April 2023"
)
```

```python
%%time
obj_plot_3d = example_objs.plot_objects(
    title = "g filter detections 09-12 April 2023",
    filters = ['g'],
    projection = "3d"
)
```
