# Object examples notebook

The Object class (not to be confused with python's object) contains data and plotting methods for an individual solar system object.

```python
import sys

sys.path.append("../Source")
```

```python
#import the object class from the ssobject module
from ssobject import Object
```

```python
# instantiate the object
example_obj = Object(mpcdesignation = "S1000zd2a", lazy_loading = False)
```

### Phase Curve

A phase curve for the object can be generated using the phase_curve method. A ScatterPlot object is returned.

phase_curve(

    start_time: Optional[float] = None, 
    end_time : Optional[float] = None,
    filters: Optional[list] = None,
    title : Optional[str] = None,
    library : Optional[str] = "matplotlib",
    cache_data: Optional[bool] = False,
    fit = None, # [None, "HG"]
    ):


```python
%%time
pc = example_obj.phase_curve(
    filters = ["g", "i", "r"], fit = "HG"
)
```

single_phase_curves is a method that generates individual phase curve plots for the specified filters. A list of ScatterPlot objects is returned.

```python
pcs = example_obj.single_phase_curves(
    filters = ["g", "r", "i", "z"],
    title = "Phase curve for S1000zd2a",
    fit = "HG",
    cache_data = True
)
```

### Light Curve

The light curve method generates a light curve within the given parameters:

light_curve(

        filters: Optional[list] = None,
        start_time : Optional[float] = None,
        end_time : Optional[float] = None,
        title : Optional[str] = None,
        time_format: Optional[Literal['ISO', 'MJD']] = 'ISO',
        library: Optional[str] = "matplotlib",
        cache_data: Optional[bool] = False
       ):

```python
%%time

example_obj.light_curve(
    title = "Light curve for all observations of object S1000zd2a"
)
```

single_light_curves provides a wrapper over the light_curve method to generate individual light curves from the specified filters:

single_light_curves(

        filters: list = ['g','r','i','z','y','u'],
        title: str = None,
        cache_data: Optional[bool] = False
    ):


```python
example_obj.single_light_curves(
    title= "Light curve for object S1000zd2a",
    filters = ["g", "r", "i", "z"],

)
```

### Plot Orbit

```python
example_obj.plot_orbit(
    projection = "2d",
    title = "Cartesian view of object S1000zd2a"
)
```

```python
example_obj.plot_orbit(
    projection = "3d",
    filters = ["g", "r", "i", "z", "y", "u"],
    title = "3d projection of observed points for object S1000zd2a"
)
```
