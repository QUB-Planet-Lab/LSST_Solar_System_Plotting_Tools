# Objects

#### The object class allows for data to be retrieved as attributes and plots to be constructed for any solar system object

```python
import sys
sys.path.append("../Source")
```

```python
from ssobject import Object
```

```python
# Provide either an mpcdesignation or ssobjectid

example_1 = Object(mpcdesignation = "S1000zd2a")
```

```python
example_1.orbital_parameters
```

```python
pc = example_1.phase_curve()

pc_1 = example_1.phase_curve(filters = ["g", "i", "r"])
# bug with "y"
```

```python
lc = example_1.light_curve(
    title = "Example 1",
    start_time = "2024-06-01",
)

lc_1 = example_1.light_curve(
    start_time = "2024-06-01",
    end_time = "2035-06-01",
    filters = ['g','r','i','z','y'],
    title = f"S100kWwNa (2024-06-01 - 2026-06-01)"
)
```

```python
example_1.plot_orbit(title = "2D orbit of object S100kWwNa")
example_1.plot_orbit(projection = "3D")
```

# Collections

```python
from collection import Collection
```

```python
example_2 = Collection(
    max_a = 1.3
)
```

```python
x = example_2.plot_objects(
    start_time = "2024-06-01",
    end_time = "2024-07-01",
    title = "Objects plot"
)
example_2.plot_objects(
    start_time = "2024-06-01",
    end_time = "2024-06-04",
    filters = ["g", "r", "i"],
    projection = "3d",
    title = "3d objects plot"
)
```

```python
example_2.orbital_param_distribution(
    "eccentricity",
     start_time = "2024-06-01",
    end_time = "2024-06-16",
    filters = ["g", "r", "i"]
)
```

```python
y = example_2.tisserand_relations(
    start_time = "2024-06-01",
    end_time = "2024-06-04",
    y = "incl",
    title = "Tisserand values across inclinations"
    #plot_type="2d_hist", "2d_hex"
)
```

```python
example_2.tisserand_relations(
    start_time = "2024-06-01",
    end_time = "2024-06-10",
    y = "incl",
    title = "Tisserand values across inclinations",
    plot_type="2d_hex" #plot_type = "2d_hist"
)
```

```python
example_2.orbital_relations(x = "incl", y = "a", start_time = "2024-06-01",
    end_time = "2024-06-10", plot_type="2d_hist", colorbar=False)
```
